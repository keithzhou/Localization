```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def calcMapping(n): # mapping for e(c) = 2 ^ n
    mapping = {}
    for i in range(1,n+1):
        mapping[i] = np.log(2**i)
        mapping[-i] = np.log(1+2**(-i))
    return mapping

def precision_cap(num, n_decimal): # a very dumb but working way to do precision cap
    s = "%" + ".%df" % n_decimal 
    return float(s % num)

def exp_fp(i,mapping,precision): # calculate exponential
    for key in reversed(sorted(mapping)):
        if i >= mapping[key]:
            multiplier = 2 ** key
            if key < 0:
                multiplier += 1.0
            return precision_cap(precision_cap(multiplier,precision) * precision_cap(exp_fp(i-mapping[key],mapping,precision),precision),precision)
    return 1.0
```

```python
print "percentage error: |exp_fixed_point(x) - e(x)| / e(x)"
ns = range(1,50,10)
xs = np.linspace(0,19,20)
precisions = [2,5,10,20,30]
plt.figure(figsize=(15,15))
for i,n_digits_after_decimal in enumerate(precisions):
    plt.subplot(3,3,i+1)
    for n in ns:
        mapping = calcMapping(n)
        plt.plot(xs, [abs(exp_fp(int(i),mapping, n_digits_after_decimal) - np.exp(i))/np.exp(i) for i in xs])
    plt.yscale('log')
    plt.legend(["n = %d"% i for i in ns], loc=0)
    plt.xlabel('input x for e(x)')
    plt.ylabel('percentage error')
    plt.title('%d digits precision' % n_digits_after_decimal)
```
