All points on a hyperbola curve have the same *difference of distances* to two fixed points. Ideally all such points form one curve. However, in practice we can only measure difference of distance up to a precision. Therefore we look at all points with difference of distance close to some target value within measurement error $\epsilon$. This $\epsilon$ prepresents accuracy on measurement of difference of distances, and it is also related to sampling rate.

Here is a setup of two microphoens located at: $M_1:(x=-10cm,y=0cm)$ and $M_2:(x=10cm,y=0cm)$. A test sound source is emitted at point $P:(x=0cm,y=10cm)$. The plot below shows the region where all points $\hat P$ satisfy: $$|(\hat P M_1 - \hat P M_2) - (P M_1 - P M_2)| < 1 cm$$ (i.e $\epsilon=1cm$. all points $\hat P$ in the region have very similar difference of distance to the two microphones compared to point $P$) 

```python
# two microphones at (x=-10cm, y=0cm) and (x=10cm, y=0cm)
pt = (0.0,0.3)
sut = sim(mics=((-0.1,0.0),(0.1,0)),sampling_resolution=0.01, grid_res=1000)
plot_hm(sut.xx, sut.yy, sut.candidates_all_paths(pt), (None, None), sut.mics, title="two microphones",dot=pt)
```

Note that the region becomes larger as you move away from the microphones. It says the same delta distance movement will generate smaller difference of distance to the microphones when you are farthur away from the microphones.

How the region changes also depends on the angle of the sound location. Plots below shows how to region looks like with angle

```python
r = 0.5
for angle in np.linspace(0,np.pi, 5):
    pt = (r * np.cos(angle),r * np.sin(angle))
    plot_hm(sut.xx, sut.yy, sut.candidates_all_paths(pt), (None, None), sut.mics, title="angle:%.2f degrees" % (angle / np.pi * 180), dot=pt)
```

Notice how regions expands as you move toward the line of the two microphones. The confusion region is smallest when difference of distances is near $0$ and largest when difference of distances is near maximum (Note: maximum difference is the distance between the two microphones)

Another variable that has imapct on the confusion region is the precision parameter $\epsilon$. For all simulation above $\epsilon=1cm$ is used. The plots below shows how the region shrinks when $\epsilon$ decreases

Note that $\epsilon=1cm$ roughly corresponds to sampling rate $f=(1.0 s)/(\frac{0.01 m}{340 m/s}) = 34KHz$. Then $\epsilon=0.1cm$ roughly corresponds to $340KHz$

```python
for sampling_resolution in (0.01,0.001):
    sut = sim(mics=((-0.1,0),(0.1,0)), sampling_resolution=sampling_resolution)
    pt = (0.5,0)
    plot_hm(sut.xx, sut.yy, sut.candidates_all_paths(pt), (None, None), sut.mics, title="epsilon:%.3f meters" % (sampling_resolution), dot=pt)
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import os.path, time
import sys
import zmq
from pylab import cm
import config
import multiprocessing
```

```python
distance = lambda p1,p2: sum((x-y) ** 2 for x,y in zip(p1,p2)) ** 0.5

class sim():
    def __init__(self,mics, sampling_resolution = 0.01, grid_res = 1000):
        self.sampling_resolution = sampling_resolution
        self.mics = mics
        self.xs = np.linspace(-1,1,grid_res)
        self.ys = np.linspace(-1,1,grid_res)
        (self.xx, self.yy) = np.meshgrid(self.xs,self.ys)

        dis = [((self.xx - x) ** 2 + (self.yy - y) **2)**0.5 for x,y in mics]
        self.dis_diffs = self.calculate_diffs(dis)
        
    def calculate_diffs(self,dis):
        result = []
        for i in range(len(dis)):
            for j in range(i+1,len(dis)):
                result.append(dis[i] - dis[j])
        return result
    
    def numerical_hyperbolas(self,f,pt):
        # pt is used to calculate distance differences
        # f is used to combine hyperbolas. use np.logic_and for intersection and np.logic_or for showing path
        dis = [distance(pt,m) for m in self.mics]
        diffs = self.calculate_diffs(dis)
        return reduce(f,(abs(a - b) < self.sampling_resolution for a,b in zip(self.dis_diffs,diffs)))
    
    def candidates_intersection(self,pt):
        return self.numerical_hyperbolas(np.logical_and,pt)
    
    def candidates_all_paths(self,pt):
        return self.numerical_hyperbolas(np.logical_or,pt)
    
    def evaluate(self,pts):
        results = []
        for pt in pts:
            cand = self.candidates_intersection(pt)
            loc = np.mean([(self.xs[x],self.ys[y]) for y, x in zip(*np.nonzero(cand))],axis=0)
            num_cand = np.sum(cand)
            results.append((loc,num_cand))
        return results
```

```python
def plot_hm(xx,yy,data = None,(mi,ma) = (None, None),dots=(),title = None, dot = None):
    fig,ax = plt.subplots()
    quad = ax.pcolormesh(xx, yy ,data , cmap=cm.RdBu, vmin=mi or np.min(data), vmax=ma or np.max(data))
    cb = fig.colorbar(quad, ax=ax)
    for x,y in dots:
        ax.plot(x,y,'oy')
    if title:
        ax.set_title(title)
    if dot:
        ax.plot(dot[0],dot[1],'xw')

def visualize_results(results,xy_real,c):
    cc = np.array([j for i,j in results])
    xy = np.array([i for i,j in results])
    ee = np.sum((xy - xy_real)**2,axis=1) ** 0.5
    print "mean error:", np.mean(ee,axis=0)
    xx, yy = np.meshgrid(np.linspace(-1,1,200), np.linspace(-1,1,200))
    plot_hm(xx,yy,cc.reshape(200,200),(None,None),c.mics,"intersection area at each point")
    #plot_hm(xx,yy,ee.reshape(200,200),(0,0.5),c.mics)
```

```python
def worker(pts):
    sut = build_sut()
    return sut.evaluate(pts)

def runInParallel(worker,pts):
    pool = multiprocessing.Pool(processes=10)
    results = pool.map(worker,chunks(pts,4000))
    return [item for sublist in results for item in sublist]
```

```python
def chunks(l, n):
    """ Yield successive n-sized chunks from l.
    """
    for i in xrange(0, len(l), n):
        yield l[i:i+n]
```
