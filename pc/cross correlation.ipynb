```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from scipy.fftpack import rfft, irfft, fftfreq
import os.path, time
import sys
```

```python
fs = 1000
t = np.linspace(0,10,10 * fs)
data = np.random.rand(len(t)) - 0.5
ch1 = data
ch2 = np.roll(data,10)
```

```python
def plot_fft(signal,fs):
    fig = plt.figure(num=None, figsize=(18, 6), dpi=80, facecolor='w', edgecolor='k')
    W = fftfreq(signal.size, 1.0 / fs)
    f_signal = rfft(signal)
    bolla = (W >= 0) & (W < 10e3)
    ax1 = fig.add_subplot(121)
    ax1.plot(signal)
    ax2 = fig.add_subplot(122)
    ax2.plot(W[bolla],f_signal[bolla])
    plt.show()
    
def pp(sig1,sig2):
    plot_fft(sig1,fs)
    plot_fft(sig2,fs)
    xcorr_output =  np.correlate(sig1,sig2,'full')
    plt.plot(xcorr_output)
    print np.argmax(xcorr_output) - (len(sig2) - 1)
    plt.show()
```

```python
pp(ch1,ch2)
```

```python
sig3 =  np.cos(2*np.pi*20*t)/10
sig32 = np.random.rand(len(t))-0.5
sig33 = np.random.rand(len(t))-0.5
```

```python
plot_fft(sig3,fs)
plot_fft(sig32,fs)
```

```python
pp(ch1+sig3,ch2+sig3)
```
